```python
name_to_letters={}
aa=open("model/apiCallsToLetters.csv")
for a in aa:
    a=a.rstrip()
    a=a.split(",")
    for item in a[1:]:
        if item!='':
            name_to_letters[item]=a[0]
```

```python
print(name_to_letters)
```

```python
import json
with open('apiDict.json') as json_file: 
    api_to_num = json.load(json_file) 
```

```python
num_to_api= {v: k for k, v in api_to_num.items()}
```

```python
import pandas as pd 
df = pd.read_csv('relevant_hashes.csv',header=None,sep='\t') 
```

```python
wf=open("encoded_family_relevant.txt","w")
for index, row in df.iterrows():
    allx=str(row[0])
    allx=allx.split(",")
    seq=allx[2]
    encoded=""
    for item in seq.split("_"):
        if item=="":
            continue
        item=int(item)
        if item in num_to_api:
            if item!="":
                name=num_to_api[item]
                if name in name_to_letters:
                    letter=name_to_letters[name]
                    encoded+=letter
    print(allx[0])
    wf.write(allx[0])
    wf.write("\t")
    print(encoded)
    wf.write(encoded)
    wf.write("\n")
    print("------")
```
